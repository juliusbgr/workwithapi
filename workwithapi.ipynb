```python
from pprint import pprint

# Make the request to the Open Notify API
url = "http://api.open-notify.org/astros.json"
response = requests.get(url)

# Convert JSON response to Python dictionary
data = response.json()

# Check the type of the variable 'data'
print("Type of data:", type(data))  # should be <class 'dict'>

# Pretty print the whole data
print("\nFull data from API:")
pprint(data)

# Print the number of people currently in space
print("\nNumber of people in space:", data['number'])

# Loop through the dictionary and print all first names
print("\nFirst names of people in space:")
for person in data['people']:
    first_name = person['name'].split()[0]  # get first name
    print(first_name)
```

```python
import os
import requests
from dotenv import load_dotenv
import pandas as pd

load_dotenv()
av_token = os.getenv('AV_TOKEN')
```

```python
# Stock Ticker Symbol
symbol = 'GOOG'

url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={symbol}&apikey={av_token}'
r = requests.get(url)
data = r.json()
data
```

```python
income_df = pd.DataFrame(data['annualReports'])
income_df
```

```python
income_df['fiscalDateEnding'] = pd.to_datetime(income_df['fiscalDateEnding'])
income_df['fiscalYear'] = income_df['fiscalDateEnding'].dt.year
cols_to_float = ['grossProfit', 'totalRevenue',
       'costOfRevenue', 'costofGoodsAndServicesSold', 'operatingIncome',
       'sellingGeneralAndAdministrative', 'researchAndDevelopment',
       'operatingExpenses', 'investmentIncomeNet', 'netInterestIncome',
       'interestIncome', 'interestExpense', 'nonInterestIncome',
       'otherNonOperatingIncome', 'depreciation',
       'depreciationAndAmortization', 'incomeBeforeTax', 'incomeTaxExpense',
       'interestAndDebtExpense', 'netIncomeFromContinuingOperations',
       'comprehensiveIncomeNetOfTax', 'ebit', 'ebitda', 'netIncome']

income_df[cols_to_float] = income_df[cols_to_float].apply(pd.to_numeric, errors='coerce')
income_df = income_df.sort_values(by='fiscalYear', ascending=True)
income_df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(10, 6))
income_df.plot(
    x='fiscalYear',
    y='totalRevenue',
    kind='bar',
    ax=ax,
    title=f'Revenue of {symbol}',
    ylabel='Revenue (Billion USD)',
    xlabel='Fiscal Year',
    legend=False
)

# Format y-axis to billions
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x/1e9:.0f}'))

plt.tight_layout()
plt.show()
```

```python
url2 = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&apikey=9WELMTZ75WHRBZUO'
r2 = requests.get(url2)
data2 = r2.json()

print(data2)
```

```python
from datetime import datetime

# Parse time series data
time_series = data2.get('Time Series (Daily)', {})

# Prepare lists for plotting
dates = []
closing_prices = []

for date_str in sorted(time_series.keys()):
    dates.append(datetime.strptime(date_str, '%Y-%m-%d'))
    closing_prices.append(float(time_series[date_str]['4. close']))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(dates, closing_prices, label='IBM Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.title('IBM Daily Closing Prices')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
